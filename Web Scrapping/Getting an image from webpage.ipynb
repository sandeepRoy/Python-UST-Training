```python
#Getting an Image from a Website
import requests
import bs4

response_data = requests.get("https://en.wikipedia.org/wiki/InSight")
soup = bs4.BeautifulSoup(response_data.text, 'lxml')

image_info = soup.select('.thumbimage')
image_info
```
